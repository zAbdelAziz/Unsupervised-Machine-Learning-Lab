The cumulant-generating function $K_X(t)$ of a random variable $X$ is defined as 
the natural logarithm of the moment-generating function
\begin{equation}
  \newcommand\E{\operatorname{\mathbb{E}}}
  \newcommand\dif{\mathrm{d}}
  K_X(t) = \log \E\left[e^{tX}\right]. 
\end{equation}
Let $\kappa_n(X) = K_X^{(n)}(0)$ the $n$-th derivative of $K_X(t)$ evaluated at $t=0$, 
then the Maclaurin series is 
\begin{equation}
  K_X(t) = \sum_{n=1}^\infty \kappa_n(X) \frac{t^n}{n!}. 
\end{equation}

<h2 style="color:rgb(0,120,170)">Task 1 (20 Points)</h2>
Show that if the random variables $X$ and $Y$ are independent then 
$\kappa_n(X+Y) = \kappa_n(X) + \kappa_n(Y)$ and $\kappa_n(\alpha X) = \alpha^n \kappa_n(X)$
holds. 

########## YOUR SOLUTION HERE ##########<br>
#### (1)
$\kappa_n(X+Y) = \frac{\mathrm{d}^n}{\mathrm{d}t^n} \big( K_{X+Y}(t) \big) \quad \bigg|\quad{t=0} \quad$ (definition of the cumulant-generating function)<br>
$= \frac{\mathrm{d}^n}{\mathrm{d}t^n} \big( \log \E\left[e^{t(X+Y)}\right] \big) \quad \bigg|\quad{t=0}$<br>
$= \frac{\mathrm{d}^n}{\mathrm{d}t^n} \big( \log \E\left[e^{tX} e^{tY}\right] \big) \quad \bigg|\quad{t=0}$<br>
$= \frac{\mathrm{d}^n}{\mathrm{d}t^n} \big( \log \E\left[e^{tX}\right] \E\left[e^{tY}\right] \big) \quad \bigg|\quad{t=0}$<br>
$= \frac{\mathrm{d}^n}{\mathrm{d}t^n} \big( \log \E\left[e^{tX}\right] + \log \E\left[e^{tY}\right] \big) \quad \bigg|\quad{t=0}$<br>
$= \kappa_n(X) + \kappa_n(Y)$<br>

#### (2)
$\kappa_n(\alpha X) = \frac{\mathrm{d}^n}{\mathrm{d}t^n} K_{\alpha X}(t) \quad \bigg|\quad{t=0}$<br>
$= \frac{\mathrm{d}^n}{\mathrm{d}t^n} \log \E\left[e^{t(\alpha X)}\right]$<br>
$= \frac{\mathrm{d}^n}{\mathrm{d}t^n} \log \E\left[e^{\alpha(tX)}\right]$<br>
$= \frac{\mathrm{d}^n}{\mathrm{d}t^n} \log \left(\E\left[e^{tX}\right]\right)^\alpha$<br>
$= \frac{\mathrm{d}^n}{\mathrm{d}t^n} \left(\alpha \log \E\left[e^{tX}\right]\right)$<br>
$= \alpha^n \frac{\mathrm{d}^n}{\mathrm{d}t^n} \log \E\left[e^{tX}\right]$<br>
$= \alpha^n \kappa_n(X)$

Show that the first four cumulants of a centered random variable $X$, i.e. $\E[X] = 0$, are 
  \begin{equation}
    \kappa_1 = 0,\quad\kappa_2 = \E[X^2],\quad\kappa_3 = \E[X^3],\quad\text{and}\quad\kappa_4 = \E[X^4] - 3(\E[X^2])^2,

########## YOUR SOLUTION HERE ##########

Taylor expansion of the moment-generating function, we have<br>
$E[e^{tX}] = \sum_{n=0}^\infty \E[X^n] \frac{t^n}{n!}$<br>
$= 1 + \E[X]t + \E[X^2]\frac{t^2}{2!} + \E[X^3]\frac{t^3}{3!} + \E[X^4]\frac{t^4}{4!} + \dots$<br>

Since $\E[X]=0$, this simplifies to<br>
$E[e^{tX}] = 1 + \E[X^2]\frac{t^2}{2!} + \E[X^3]\frac{t^3}{3!} + \E[X^4]\frac{t^4}{4!} + \dots$<br>

Taylor expansion of $\log(1+a) \implies log(1+a) = a - \frac{a^2}{2} + \frac{a^3}{3} - \frac{a^4}{4} + \dots$<br>
Let $a = \E[e^{tX}] - 1$, we get<br>
$\log\E[e^{tX}] = \E[e^{tX}] - 1 - \frac{1}{2}(\E[e^{tX}] - 1)^2 + \frac{1}{3}(\E[e^{tX}] - 1)^3 - \frac{1}{4}(\E[e^{tX}] - 1)^4 + \dots$<br>

Simplify, drop all terms of order higher than 4. we get<br>
$\log\E[e^{tX}] = \frac{1}{2}\E[X^2]t^2 + \frac{1}{3}\E[X^3]t^3 + \frac{1}{4}\left(\E[X^4] - (\E[X^2])^2\right)t^4 + \dots$<br>

Teke derivative w.r.t $t$ and evaluating at $t=0$, we get<br>
$\kappa_1 = 0$<br>
$\kappa_2 = \E[X^2]$<br>
$\kappa_3 = \E[X^3]$<br>
$\kappa_4 = \E[X^4] - 3(\E[X^2])^2$<br>

Consider a standard normal random variable $X \sim p(x) = \frac{1}{\sqrt{2\pi}} e^{-\frac{x^2}{2}}$. Show that $\E[X] = 0$ and $\E[X^n] = (n-1) \E[X^{n-2}]$ for $n \geq 2$ and use this to determine the third and fourth cumulants of $X$. *Hint:* Use the fact that $\int f'(x) e^{f(x)} \dif x = e^{f(x)}$. 

########## YOUR SOLUTION HERE ##########<br>
The first two moments of $X$:<br>
##### (1) First Moment:
$\E[X] = \int_{-\infty}^\infty x p(x) \dif x$<br>
$= \int_{-\infty}^\infty x \frac{1}{\sqrt{2\pi}} e^{-\frac{x^2}{2}} \dif x = 0 $

##### (2) Second Moment:
$\E[X^2] = \int_{-\infty}^\infty x^2 p(x) \dif x$<br>
$= \int_{-\infty}^\infty x^2 \frac{1}{\sqrt{2\pi}} e^{-\frac{x^2}{2}} \dif x = 1$<br>

##### Higher (n) Moments:
Since $p(x)$ is even and by using integration by parts we get:<br>
$\E[X^n] = \int_{-\infty}^\infty x^n p(x) \dif x $<br>
$= \int_{-\infty}^\infty x^{n-1} \cdot x p(x) \dif x$<br>
$= \int_{-\infty}^\infty x^{n-1} \cdot \frac{1}{\sqrt{2\pi}} (-x) \frac{\dif}{\dif x} e^{-\frac{x^2}{2}} \dif x $<br>
$= \frac{1}{\sqrt{2\pi}} \left[ -x^{n-1} e^{-\frac{x^2}{2}} \right]{-\infty}^\infty + \frac{n-1}{\sqrt{2\pi}} \int{-\infty}^\infty x^{n-2} e^{-\frac{x^2}{2}} \dif x$<br>
$= 0 + (n-1) \E[X^{n-2}]$

Using this formula, we can compute the third and fourth cumulants of $X$:

$\kappa_3(X) = \E[X^3] - 3 \E[X] \E[X^2] = \E[X^3] = (3-1) \E[X] = 0 \implies$ Skewness of X = 0<br>
$\kappa_4(X) = \E[X^4] - 3 (\E[X^2])^2 = 3 \E[X^2] - 3 (\E[X^2])^2 = 3 (1) - 3 (1)^2 = 0\implies$ Excess Kurtosis of X = 0<br>

Consider the cocktail party problem. Read the left channel of the files `birds.wav`, `ocean.wav`, `frogs.wav` using the `scipy.io.wavfile` package. Then construct a full-rank $3 \times 3$ mixing matrix and produce 3 different mixings of those signals. You can listen to them using the `sounddevice` package. Finally, use `FastICA` from the `sklearn.decomposition` package to recover the original sources. Listen to the recovered sources and compare them to the originals to check if ICA was successful. *Hint:* Scale the signals to $[-1,1]$ before playing them. 

```python
import scipy.io.wavfile
import sounddevice as sd

########## YOUR SOLUTION HERE ##########
def get_file(fname):
    # Open File
    rate, mat = scipy.io.wavfile.read(fname)
    # Select Left
    mat_left = mat[:, 0]
    # Scale [-1, 1]
    mat_left /= np.max(np.abs(mat_left),axis=0)
    return mat_left, rate

def make_rank_mat(size=(3,3)):
    while True:
        A = np.random.rand(*size)
        if np.linalg.det(A) != 0:
            break
    return A
            
            
birds, birds_rate = get_file('birds.wav')
ocean, ocean_rate = get_file('ocean.wav')
frogs, frogs_rate = get_file('frogs.wav')
# print(birds.shape, birds_rate)

full_rank = make_rank_mat()


# Mix Signals
stacked = np.vstack((birds, ocean, frogs)).T
# mixed = np.dot(full_rank, np.array([frogs, birds, ocean]))
mixed = stacked @ full_rank
# print(mixed.shape)


sd.play(mixed[:,0], birds_rate)
sd.stop()
sd.wait()

sd.play(mixed[:,1], birds_rate)
sd.stop()
sd.wait()

sd.play(mixed[:,2], birds_rate)
sd.wait()
sd.stop()

# ICA
ica = FastICA(n_components=3)
# Compute Recovered
recovered = ica.fit_transform(mixed)
# Scale Recovered
recovered = recovered / np.max(np.abs(recovered))

# Birds

sd.play(recovered[:,0], birds_rate)
sd.wait()
sd.stop()
# Ocean

sd.play(recovered[:,1], birds_rate)
sd.wait()
sd.stop()
# Frogs

sd.play(recovered[:,2], birds_rate)
sd.wait()
sd.stop()


fig, axs = plt.subplots(2, 3, figsize=(16, 5))
axs[0, 0].plot(mixed[:,0])
axs[0, 0].set_title('Original')
axs[0, 1].plot(mixed[:,1])
axs[0, 1].set_title('Original')
axs[0, 2].plot(mixed[:,2])
axs[0, 2].set_title('Original')
axs[1, 0].plot(recovered[:, 0])
axs[1, 0].set_title('Recovered')
axs[1, 1].plot(recovered[:, 1])
axs[1, 1].set_title('Recovered')
axs[1, 2].plot(recovered[:, 2])
axs[1, 2].set_title('Recovered')
plt.tight_layout()
plt.show()

# The Samples are indistinguishable from each other which implies the ica was successful to some extent;
# yet, when plotting the graph of the recovered samples there's a slight difference between the the magnitude of some areas -> loss of information
```
